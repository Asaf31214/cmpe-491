```python
import random
from os import path

import numpy as np
import polars as pl
import torch
from IPython.display import display
from sklearn.preprocessing import StandardScaler
from torch import nn, optim
```

```python
SEED = 491
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
data_path = "../../data/previous"
seer_path = path.join(data_path, "processed_seer_with_age.csv")
cohort_path = path.join(data_path, "processed_cohort_with_age.csv")
seer = pl.read_csv(seer_path)
cohort = pl.read_csv(cohort_path)
```

```python
label_col = "target"
cols = seer.select(pl.exclude(label_col)).columns
X_seer = seer.select(pl.exclude(label_col)).to_numpy().astype(np.float32)
y_seer = seer.get_column(label_col).to_numpy().astype(np.int64).ravel()
X_cohort = cohort.select(pl.exclude(label_col)).to_numpy().astype(np.float32)
y_cohort = cohort.get_column(label_col).to_numpy().astype(np.int64).ravel()
```

```python
scaler_seer = StandardScaler()
scaler_cohort = StandardScaler()
X_seer_std = scaler_seer.fit_transform(X_seer).astype(np.float32)
X_cohort_std = scaler_cohort.fit_transform(X_cohort).astype(np.float32)
```

```python
input_dim = X_seer_std.shape[1]
cond_dim = 1
latent_dim = 24
hidden = 128


class CVAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.enc1 = nn.Linear(input_dim + cond_dim, hidden)
        self.enc2 = nn.Linear(hidden, hidden)
        self.mu = nn.Linear(hidden, latent_dim)
        self.logvar = nn.Linear(hidden, latent_dim)
        self.dec1 = nn.Linear(latent_dim + cond_dim, hidden)
        self.dec2 = nn.Linear(hidden, hidden)
        self.out = nn.Linear(hidden, input_dim)

    def encode(self, x, c):
        h = torch.relu(self.enc1(torch.cat([x, c], 1)))
        h = torch.relu(self.enc2(h))
        return self.mu(h), self.logvar(h)

    def reparam(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z, c):
        h = torch.relu(self.dec1(torch.cat([z, c], 1)))
        h = torch.relu(self.dec2(h))
        return self.out(h)

    def forward(self, x, c):
        mu, logvar = self.encode(x, c)
        z = self.reparam(mu, logvar)
        x_hat = self.decode(z, c)
        return x_hat, mu, logvar
```

```python
model = CVAE().to(device)
mse = nn.MSELoss(reduction="mean")
```

```python
seer_optimizer = optim.Adam(model.parameters(), lr=1e-3)
seer_inputs = torch.tensor(X_seer_std, device=device)
seer_labels = torch.tensor(y_seer, dtype=torch.float32, device=device).unsqueeze(1)

for epoch in range(60):
    seer_recon, seer_mu, seer_logvar = model(seer_inputs, seer_labels)
    seer_recon_loss = mse(seer_recon, seer_inputs)
    seer_kld_loss = -0.5 * torch.mean(1 + seer_logvar - seer_mu.pow(2) - seer_logvar.exp())
    seer_loss = seer_recon_loss + 0.01 * seer_kld_loss
    seer_optimizer.zero_grad()
    seer_loss.backward()
    seer_optimizer.step()
```

```python
cohort_optimizer = optim.Adam(model.parameters(), lr=5e-4)
cohort_inputs = torch.tensor(X_cohort_std, device=device)
cohort_labels = torch.tensor(y_cohort, dtype=torch.float32, device=device).unsqueeze(1)

for epoch in range(5):
    cohort_recon, cohort_mu, cohort_logvar = model(cohort_inputs, cohort_labels)
    cohort_recon_loss = mse(cohort_recon, cohort_inputs)
    cohort_kld_loss = -0.5 * torch.mean(1 + cohort_logvar - cohort_mu.pow(2) - cohort_logvar.exp())
    cohort_loss = cohort_recon_loss + 0.005 * cohort_kld_loss
    cohort_optimizer.zero_grad()
    cohort_loss.backward()
    cohort_optimizer.step()
```

```python
target_pos_rate = 0.25
n_syn = 5000
y_syn = np.random.binomial(1, target_pos_rate, size=n_syn).astype(np.int64)
with torch.no_grad():
    z = torch.randn(n_syn, latent_dim, device=device)
    c = torch.tensor(y_syn, dtype=torch.float32, device=device).unsqueeze(1)
    x_gen_std = model.decode(z, c).cpu().numpy().astype(np.float32)
X_syn = scaler_cohort.inverse_transform(x_gen_std)
```

```python
neg_idx = np.where(y_syn == 0)[0]
pos_idx = np.where(y_syn == 1)[0]
real_neg = X_cohort[y_cohort == 0]
real_pos = X_cohort[y_cohort == 1]

real_neg_mean = real_neg.mean(axis=0)
real_neg_std = real_neg.std(axis=0)
syn_neg_mean = X_syn[neg_idx].mean(axis=0)
syn_neg_std = X_syn[neg_idx].std(axis=0)
X_syn[neg_idx] = (X_syn[neg_idx] - syn_neg_mean) / (syn_neg_std + 1e-8) * real_neg_std + real_neg_mean

real_pos_mean = real_pos.mean(axis=0)
real_pos_std = real_pos.std(axis=0)
syn_pos_mean = X_syn[pos_idx].mean(axis=0)
syn_pos_std = X_syn[pos_idx].std(axis=0)
X_syn[pos_idx] = (X_syn[pos_idx] - syn_pos_mean) / (syn_pos_std + 1e-8) * real_pos_std + real_pos_mean
```

```python
syn_df = pl.DataFrame(X_syn, schema=cols).with_columns(pl.Series(label_col, y_syn))
out_path = path.join(data_path, "synthetic_patients_1.0.csv")
syn_df.write_csv(out_path)
```

```python
real_m = X_cohort.mean(axis=0)
syn_m = X_syn.mean(axis=0)
real_s = X_cohort.std(axis=0)
syn_s = X_syn.std(axis=0)
```

```python
tbl = pl.DataFrame(
    {
        "col": cols,
        "real_mean": real_m.astype(np.float32),
        "syn_mean": syn_m.astype(np.float32),
        "real_std": real_s.astype(np.float32),
        "syn_std": syn_s.astype(np.float32),
    }
)
display(tbl)
```
